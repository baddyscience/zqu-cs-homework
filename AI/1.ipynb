```python
import random
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['font.sans-serif'] = ['SimHei']  # 设置字体为SimHei
plt.rcParams['axes.unicode_minus'] = False  # 解决保存图像是负号'-'显示为方块的问题

# 模拟掷色子700次
num_rolls = 700
rolls = [random.randint(1, 6) for _ in range(num_rolls)]

# 统计各个点数出现的频次和比例
counts = np.bincount(rolls)[1:]
proportions = counts / num_rolls

# 打印频次和比例
for i in range(6):
    print(f"点数 {i+1}: 频次 = {counts[i]}, 比例 = {proportions[i]:.4f}")

# 可视化结果
plt.figure(figsize=(8, 6))
sns.barplot(x=np.arange(1, 7), y=counts, palette="viridis")
plt.title("掷色子700次的结果")
plt.xlabel("点数")
plt.ylabel("频次")
plt.show()
```
